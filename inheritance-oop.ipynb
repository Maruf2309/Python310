# Interitance- OOP

* Inheritance is a process by which a class characteristic transferred to another class
* Example- Mom & Daughter: Some properties of mom inherits by her daughter

### Excercise:1

```python
# Inheritance 
class myself:                         # Parent class
    def text1(self):
        print("I am Maruf Hossain")
        
    def text2(self):
        print("I like Python")
        
    def text3(self):
        ("I am an AI enthusiast")
        
class pythonlan:                        # Clild class
    def text4(self):
        print("Programming is really fun, love it!")
```

```python
# Calling second class -- Here second class just can access its associated module i.g text4, not can access text1-text3
selfintro = pythonlan()
selfintro.text4()
```

```python
# Calling second class -- It will return error, because the module/object associated by parent class
selfintro = pythonlan()
selfintro.text1()
selfintro.text2()
selfintro.text2()
```

#### Interpretation

1. Here there are two class:
2. Parent class/ Super Class:  ( myself)
3. Clild Class / Subclass / derived class :  pythonlan
4. Intering into clild class, we can not access into "Parent Class" for that need to inheritance of that first!

* Parent class characteristic can be passed into child class
* Child class characteristics can not be passed into "Parent Class"
* For passing parent class we need to make a class child/ subclass explicitely!

* Parent class can not take anything from child class,but child class can take something from parent class

```python
# Inheritance - Making a class child (Inheritance application)

class myself:
    def text1(self):
        print("I am Maruf Hossain")
        
    def text2(self):
        print("I like Python")
        
    def text3(self):
        print("I am passionate for AI")
        
        
class pythonlan(myself):      # Inheritance child class with parent class
    def text4(self):
        print("Programming is really fun, love it!")
        

# Object calling --- After inheritance, we can access parent class
selfintro = pythonlan()
selfintro.text1()
selfintro.text2()
selfintro.text3()
selfintro.text4()
```

# Excercise - 2 Method Overriding

#### Method Overriding

Method overriding is an ability of any object-oriented programming language that allows a subclass or child class to provide a specific implementation of a method that is already provided by one of its super-classes or parent classes. When a method in a subclass has the same name, same parameters or signature and same return type(or sub-type) as a method in its super-class, then the method in the subclass is said to override the method in the super-class.

* Every class must have its 'own' function, which is called <b> init function </b>

#### Excercise - 2

```python
# Method Overriding
class myself:
    def __init__(self):
        print("Statement A: From Parent Class")
        
class pythonlan(myself):
    def __init__(self):
        print("Statement B: From Child Class")

# Calling
result = pythonlan()
```

#### Narration

1. That can be a MCQ, why Statement A will not be returned, there can be a story like this:
2. Exmample: Your father having a house & you have a house too, but regardless your father's house, child must belong to his own house, so it statement B printed
3. In nutshell, print will be as per own class association. First own print the other
4. But, if we want, then we can also return "Statement A" which is inspired by <b> Method Overriding </b>, Using super class follow test 2.2

```python
## Test 2.2 ( Method Overriding) - Single Level Inheritance
class myself:
    def __init__(self):
        print("Statement A: From Parent Class")
        
class pythonlan(myself):
    def __init__(myself):
        super().__init__()
        print("Statement B: From Child Class, both retured due to using super().__init__(). This is concept of Method Overriding!")

# Calling
result = pythonlan()
```

```python
## Test 2.3 ( Method Overriding)- Super class order can be altered & that will print as per the class order
class myself:
    def __init__(self):
        print("Statement A: From Parent Class")
        
class pythonlan(myself):
    def __init__(myself):
        print("Statement B: From Child Class, both retured due to using super().__init__(). This is concept of Method Overriding!")
        super().__init__()     # Alter super class position

# Calling
result = pythonlan()
```

### Types of Inheritance

1. Sigle Level Inheritance
2. Multi Level Inheritance
3. Multiple Inheritance

```python
# Multi Level Interitance

class myself:                   # Parent Class-1
    def text1(self):
        print("Parent Class 1")
        
class myself1(myself):         # Parent Class-2
    def text2(self):
        print("Child Class of myself")        
        
class pythonlan(myself1):
    def __init__(self):
        super().text1()
        super().text2()
        print("Pythonlan is child class of myself1. This is example of Multi Level Inheritance, where nested subclass one after one!")
        
# Object calling
selfintro = pythonlan()
```

* Here <b>Myself1</b> is considered to be subclass of <b>Myself</b>, similarly <b>pythonlan</b> is considered to be subclass of myself1

```python
# Multiple Interitance (Multiple Parent Classes)

class parent_class1:                   # Parent Class-1
    def text1(self):
        print("Parent Class 1")
        
class parent_class2:                 # As like multi level no need to pass first class inside this 2nd parent class
    def text2(self):
        print("Parent Class 2")        
        
class pythonlan(parent_class1,parent_class2): # Directly pass multiple parent classes with comma inside parenthesis
    def __init__(self):
        super().text1()
        super().text2()
        print("This is example of Multiple Inheritance, where multiple parent classes called togehter with Comma!, Both classes called in a single class")
        
# Object calling
selfintro = pythonlan()
```

* parent_class1: Like Father
* parent_class2: Like Mother
* Sub class (pytholan): Like sons
