Example 1: Adding Methods to the Student Class

```python
class Student:
    def __init__(self, roll, gpa):
        self.roll = roll
        self.gpa = gpa 

    def display_info(self):
        print(f"Roll: {self.roll}, GPA: {self.gpa}")

# Object Creation
reza = Student(10, 3.25)
mitu = Student(12, 4.0)

# Calling the method
reza.display_info()
mitu.display_info()
```

Example 2: Adding Multiple Attributes and Methods

```python
class Student:
    def __init__(self, name, roll, gpa, major):
        self.name = name
        self.roll = roll
        self.gpa = gpa
        self.major = major

    def display_info(self):
        print(f"Name: {self.name}, Roll: {self.roll}, GPA: {self.gpa}, Major: {self.major}")

    def is_honors(self):
        return self.gpa >= 3.5

# Object Creation
reza = Student("Reza", 10, 4.0, "Computer Science")
mitu = Student("Mitu", 15, 3.40, "Physics")

# Calling methods
reza.display_info()
print(f"Is Reza an honors student? {'Yes' if reza.is_honors() else 'No'}")

mitu.display_info()
print(f"Is Mitu an honors student? {'Yes' if mitu.is_honors() else 'No'}")
```

```python
# Test 3

class Animal:
    def get_value(self, name, age, food, color):
        # Attribute/ Instance
        self.name = name
        self.age = age
        self.food = food 
        self.color = color

    # Method Define for print
    def printall(self):
        print(f"Name: {self.name},Age: {self.age}, Food:{self.food}, Color:{self.color}")

# Object Creation (Out of Indentation)
Tiger = Animal()                                   # We can Initialize object (value assing ) with constructor approach
Tiger.get_value('Tiger',5, 'Meat', 'Yellow')       # This can be avoided while applied constructor- code reduce
Tiger.printall()

# Object 2
Cat = Animal()
Cat.get_value('Cat',3, 'Fish','Black')
Cat.printall()

# Object 3
Dog = Animal()
Dog.get_value('Dog',4, 'Chicken','White')
Dog.printall()
```

#### Constructor

```python
class animal:
    def __init__(self, name, age, food, color) -> None:            # init is for define constructor
        self.name = name 
        self.age = age 
        self.food = food 
        self.color = color 

    # Define method for print
    def display(self):
        print(f"Name : {self.name}, Age: {self.age}, Food: {self.food}, Color: {self.color}")

# Object Define
Cat = animal('Cat', 3, 'Fish','Black')
Cat.display()

# Object 2
Dog = animal('Dog', 5, 'Meat','White')
Dog.display()

# Object 3 
Cow = animal('Cow', 4, 'Grass','Red')
Cow.display()
```
