**Types of Inheritances**
1. Hierarchical Inheritance
2. Multi-Level Inheritance
3. Muliple Inheritance


**Hierarchical Inheritance**
- A particular `class` be inheritanced by multiple `classes`. Example `Shape` class be inheritance by `Triangle` and `Rectangle` classes


**Multi-Level Inheritance**
- `A` class inheritnace by `B` & `B` inheritanced by `C`. Its `Chronological`. One after one

# Implementation

```python
class A:
    def display1(self):
        print("I am inside A class")

class B:
    def display2(self):
        print("I am inside B class")

class C: 
    def display3(self):
        print("I am inside C class")
```

```python
# Multi- Level 
class A:
    def display1(self):
        print("I am inside A class")

class B(A): # Inherits all attributes (display1) from A
    def display2(self):
        print("I am inside B class")

        ## B Class has two method display1, display 2(display2 associated with B itself)

class C(B): # Inherits all attributes (display2) from B
    def display3(self):
        print("I am inside C class")

# Creating Object for Class C
ob1 = C()
ob1.display3()    # All Methods will be visible after . operator due to `Multi-Level` Inheriance

# In C class One method was created, but due to multi-level inheriance now it can assess all the 3 classes
```

**How can call by display3 method**

```python
# Multi-Level Inhertitance with Supper

class A :
    def display1(self):
        print("I am inside Class A")

class B(A): 
    def display2(self):
        print("I am inside Class B")

class C(B):
    def display3(self):
        super().display1()  # method3 will print all display by using super
        super().display2()
        print("I am inside Class C")

# Object Creation
ob1 = C()

ob1.display3()          # Display3 will print all due to using Super

```
