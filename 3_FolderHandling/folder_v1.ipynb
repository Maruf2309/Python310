# Working through directory

- There is a ways by which we can iterate throught all the path inside a directory

```python
os.listdir() # this is a high level info, that mean its shows list of files & folders in current dirs, but we dont know what stays inside folders, we can!
```

```python
os.walk('.')  # . mean current dir, it returns generator path / iterable path, return generator object on which we can iterate
```

```python
# we can iteratate over walk
for dirpath, dirnames, filenames in os.walk('.'):
    print(dirpath)   # let say print path
```

```python
# we can iteratate over walk
for dirpath, dirnames, filenames in os.walk('.'):
    print(dirnames) 
```

```python
# we can iteratate over walk
for dirpath, dirnames, filenames in os.walk('.'):
    print(filenames) 
```

```python
# we can iteratate over walk
for dirpath, dirnames, filenames in os.walk('.'):
    print(dirpath, dirnames, filenames)  # path >> dirname >> filename
```
